```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df_2 is your DataFrame
plt.figure(figsize=(15, 10))  # Set the size of the heatmap

# Convert categorical columns to numerical using one-hot encoding (get_dummies)
categorical_columns = ['Biz_Travel', 'Dept', 'Edu_Field', 'Gender', 'JobRol', 'Marital_Status', 'Over_Time']
df_encoded = pd.get_dummies(df_2, columns=categorical_columns)

# Calculate the correlation matrix
corr_col = df_encoded.corr().round(2)

# Create a heatmap
sns.heatmap(data=corr_col, center=0, vmin=-1, vmax=1, annot=True, cmap='Blues')

# Add labels, titles, etc. as needed
plt.title('Correlation Heatmap')
plt.show()

```
